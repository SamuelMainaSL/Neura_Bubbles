## Compile JSON File with Models UIDs and Attributes

```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_excel("/home/samuel/Downloads/Bubble_UID.xlsx")
df
```

```python
drop_columns = [colm for colm in df.columns if "Unnamed" in colm]


df.drop(drop_columns, axis=1, inplace=True)
df
```

```python
df["UID_S.No"].isna().sum()
```

```python
new_df = df[pd.notna(df["Bubble ID"])]
```

```python
new_df["UID_S.No"].isna().sum()
```

```python
new_df["Shipping Status (from Production Unit India)"].replace({np.nan: False, "Shipped": True}, inplace=True)
new_df["Bubble Status"].replace({np.nan: False, "x": True, "O":False}, inplace=True)
new_df.fillna("", inplace=True)
new_df.drop(["UID_S.No", "DataSet"], axis=1, inplace=True)
new_df
```

```python
dynamo_db_items = []
for idx in range(100):
    dynamo_db_items.append({
        "UID": new_df["UID"].iloc[idx],
        "ID": new_df["Bubble ID"].iloc[idx],
        "shippedStatus": bool(new_df["Shipping Status (from Production Unit India)"].iloc[idx]),
        "destinationShipped": new_df["Shipping Location"].iloc[idx],
        "usageCount": 0,
        "orderStatus": str(new_df["Order Status"].iloc[idx]).replace("Recieved", "Received"),
        "raptured": bool(new_df["Bubble Status"].iloc[idx]),
        "usages": [
            {
                "timestamp" : 0,
                "systemID": "",
                "userEmailID": "",
                "location": ""
            }
        ]
    })
    
dynamo_db_items
```
