## Compile JSON File with Models UIDs and Attributes

```python
import pandas as pd
import json
```

```python
df = pd.read_excel("/home/samuel/Zoho WorkDrive (SurgeonsLab AG)/My Folders/Canisters_and_Models_Counterfeit_Measures/Bubble_UID.xlsx")
df
```

```python
dictionary = dict(zip(df["UID"][:100], df["Bubble ID"][:100]))
```

```python
df["UID_S.No"].isna().sum()
```

```python
new_df = df[pd.notna(df["UID_S.No"])]
```

```python
new_df["UID_S.No"].isna().sum()
```

```python
dictionary = dict(zip(new_df["UID"][:100], new_df["Bubble ID"][:100]))
```

```python
comprehensive_dict = {}

for key in dictionary:
    comprehensive_dict[key] = {
        "ID": dictionary[key],
        "shippedStatus": True,
        "destinationShipped": "",
        "usageCount": 0,
        "raptured": False,
        "usage": [
            {
                "timestamp" : 0,
                "systemID": "",
                "userEmailID": "",
                "location": ""
            }
        ]
    }
```

```python
with open("data/models_iud_id.json", "w") as outfile:
    json.dump(comprehensive_dict, outfile, indent=4, separators=(',', ': '))
```
